```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv('student_marks.csv')
df
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(20,10))
plt.bar(x=df["Student_ID"],
       height = df["Test_4"],
       color="midnightblue")
plt.yticks(fontsize=13)
plt.title("Student marks ",fontsize=20, fontweight="bold")
plt.xticks(rotation = 45, fontsize = 13)
plt.ylabel("student marks ",fontsize= 20)
plt.show()
plt.savefig("student_marks.png")
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize=(20,10))
plt.pie(df['Student_ID'],
       labels=df['Test_5'].values,
       autopct ='%.2f%%',
       textprops={'size':'x-large',
                 'fontweight':'bold'})

plt.show()
```

```python
colors= ["#011638", "#7e2987", "ef2026"]
labels = ["Test_1", "Test_2","Test_3"]
plt.figure(figsize=(12,6))
plt.stackplot(df["Student_ID"],
             df["Test_1"],
             df["Test_2"],
             df["Test_3"],
             df["Test_4"],
             colors=colors,
             edgecolor = 'none')
plt.legend(labels=labels, loc = "upper left")
plt.ylabel("students marks",fontsize=13)
plt.title("students marks ",fontsize = 15, weight = "bold")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.hist(df["Test_2"],
        bins=8,
        color="#108A99")
plt.title("Student marks ",fontsize=15,weight="bold")
plt.xlabel("Test_marks",fontsize=15,weight="bold")
sns.despine()
plt.show()
```
