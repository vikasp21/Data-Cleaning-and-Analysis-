```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import statsmodels.api as sm
import sklearn
sns.set()
```

```python
data = pd.read_excel("BDA.xlsx")
data
```

```python
data.describe()
```

```python
y=data['DSA']
x1=data['NETWORK']
```

```python
plt.scatter(x1,y)
plt.xlabel('NETWORK',fontsize=20)
plt.ylabel('DSA',fontsize=20)
plt.show()
```

```python
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat=-0.1986*x1+112.7497
fig=plt.plot(x1,yhat,lw=4,c='orange',label='regression line')
plt.xlabel('NETWORK',fontsize=20)
plt.ylabel('DSA',fontsize=20)
plt.show()
```
