```python
plt.figure(figsize=(15,13))
plt.bar(x=data["Country"],
       height=data["Monthly Revenue"],
       color="midnightblue")
plt.xticks(rotation=45,fontsize=14)
plt.yticks(fontsize=13)
plt.title("Netflix Visualization",fontsize=15,weight="bold")
plt.ylabel("dataset on monthly revenue",fontsize=15,weight="bold")
plt.show()
```
